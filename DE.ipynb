```python
import pdfplumber
import pandas as pd


with pdfplumber.open("/content/UP.pdf") as pdf:
    all_pages_data = []


    for page in pdf.pages:
        table = page.extract_table()
        if table:
            all_pages_data.extend(table)

if all_pages_data:
    df = pd.DataFrame(all_pages_data[1:], columns=all_pages_data[0])
    df.to_excel("file.xlsx", index=False)
else:
    print("No table data found in the PDF.")
```

```python
ds.columns = [
    'Sr_No', 'Name', 'Category', 'Full_Address', 'District',
    'Type_of_Unit', 'Financing_Bank', 'Financing_Branch', 'Year_of_Sanction',
    'Capacity_MT', 'Project_Cost_Rs_Lakh', 'Eligible_Subsidy',
    'Advance_Subsidy_Rs_Lakh', 'Final_Subsidy_Rs_Lakh', 'Total_Subsidy_Rs_Lakh'
]
```

```python
columns_to_drop = ['Sr_No', 'Category', 'Financing_Bank', 'Financing_Branch',
                   'Eligible_Subsidy', 'Advance_Subsidy_Rs_Lakh',
                   'Final_Subsidy_Rs_Lakh', 'Total_Subsidy_Rs_Lakh']
data_cleaned = ds.drop(columns=columns_to_drop)
```

```python
data_cleaned = data_cleaned.drop_duplicates(subset='Name', keep='first')
```

```python
data_cleaned['Capacity_MT'] = pd.to_numeric(data_cleaned['Capacity_MT'], errors='coerce')
data_cleaned['Year_of_Sanction'] = pd.to_datetime(data_cleaned['Year_of_Sanction'], errors='coerce')
```

```python
data_cleaned['Full_Address'] = data_cleaned['Full_Address'].fillna('Unknown')
data_cleaned['District'] = data_cleaned['District'].fillna('Unknown')
```

```python
data_cleaned.to_excel('cleaned_cold_storage_data.xlsx', index=False)
```

```python
from geopy.geocoders import Nominatim
import pandas as pd
import time

geolocator = Nominatim(user_agent="cold_storage_locator")


def geocode_address(address):
    try:
        location = geolocator.geocode(address)
        return location.latitude, location.longitude
    except:
        return None, None


data_cleaned['Latitude'], data_cleaned['Longitude'] = zip(*data_cleaned['Full_Address'].apply(geocode_address))
data_cleaned.to_excel('geocoded_cold_storage_data.xlsx', index=False)
```

```python
data_cleaned
```

```python
import folium

map_center = [data_cleaned['Latitude'].mean(), data_cleaned['Longitude'].mean()]
m = folium.Map(location=map_center, zoom_start=6)

for i, row in data_cleaned.iterrows():
    if pd.notna(row['Latitude']) and pd.notna(row['Longitude']):
        folium.Marker(
            location=[row['Latitude'], row['Longitude']],
            popup=f"{row['Name']} ({row['Capacity_MT']} MT)",
        ).add_to(m)
m.save('cold_storage_map.html')
```

# Capacity Distribution: Using Different Marker Colors or Sizes

```python

m = folium.Map(location=map_center, zoom_start=6)

def get_color(capacity):
    if capacity < 5000:
        return 'green'
    elif 5000 <= capacity < 10000:
        return 'orange'
    else:
        return 'red'

for i, row in data_cleaned.iterrows():
    if pd.notna(row['Latitude']) and pd.notna(row['Longitude']):
        folium.CircleMarker(
            location=[row['Latitude'], row['Longitude']],
            radius=row['Capacity_MT'] / 1000,
            color=get_color(row['Capacity_MT']),
            fill=True,
            fill_color=get_color(row['Capacity_MT']),
            popup=f"{row['Name']} ({row['Capacity_MT']} MT)",
        ).add_to(m)

m.save('cold_storage_capacity_map.html')
```

# Cluster Analysis: Identifying Regions with High Concentration of Cold Storages

```python
import folium
from folium.plugins import MarkerCluster

map_center = [data_cleaned['Latitude'].mean(), data_cleaned['Longitude'].mean()]
m = folium.Map(location=map_center, zoom_start=6)
marker_cluster = MarkerCluster().add_to(m)

for i, row in data_cleaned.iterrows():
    if pd.notna(row['Latitude']) and pd.notna(row['Longitude']):
        folium.Marker(
            location=[row['Latitude'], row['Longitude']],
            popup=f"{row['Name']} ({row['Capacity_MT']} MT)",
        ).add_to(marker_cluster)
m.save('cold_storage_cluster_map.html')
```
